```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('exams.csv')
# importujem csv fajl
```

```python
df.head()
# prikazujem sve kroz tabelu
```

```python
df.info()
# ispisuje tip podataka , string(object) ili kao integer (broj)
```

plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'division', palette = 'viridis', size = 'work experience')

```python
df['math score'].unique()
# izbacuje sve unique podatke, unikate
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'math score', y = 'reading score', data = df, hue = 'race/ethnicity', palette = 'viridis', s = 200, alpha = 0.5)
# prikazuje relaciju slicnost  izmedju dva podatka , broja.
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'math score', y = 'reading score', data = df, style = 'parental level of education', hue = 'writing score', size = 'reading score')
# Prikazuje statistiku
```

```python
sns.rugplot(x = 'writing score', data = df, height = 0.5)
# Prikazuje grafikon najboljih opoena
```

```python
sns.displot(data = df, x = 'writing score', kde = True)
# prikaz podskupova podataka
```

```python
sns.histplot(data = df, x = 'writing score', bins = 100)
# Najbolji rezlutat 
```

```python
plt.figure(figsize = (10, 4), dpi = 200)
sns.countplot(x = 'math score', data = df)
# Prikaz poena detaljni prikaz
```

```python
plt.figure(figsize = (10, 4), dpi = 200)
sns.countplot(x = 'parental level of education', data = df, hue = 'test preparation course', palette = 'Paired')
# Prikaz koji su polozili kurs plava boja i  koji nisu belicasto plava.
```
